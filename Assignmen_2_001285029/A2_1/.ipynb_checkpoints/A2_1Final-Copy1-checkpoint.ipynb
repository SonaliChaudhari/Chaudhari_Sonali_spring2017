## All the required imports

```python
import glob
import nltk
import string
import csv
from nltk.corpus import stopwords
```

## Use of 'glob' ’ to open and read files (Guternberg Files)

```python
#all_files =[]             # List to store the names of all the files 
#print ('All FILE NAME:')  # To print the statment

#Retrieving all the files with .txt extension from the given path
file = []
for name in glob.glob('/Users/sonalichaudhari/nltk_data/corpora/gutenberg/*.txt'):  
    with open(name, 'r', encoding = 'ISO-8859-1') as infile:
        file.append(infile.read())     # Appending every file data that is being read

#To print all the files contents 
#print(file) 
```

## Removing stopwords, punctuations and digits attached with individual words

```python

file =[''.join(c for c in s if c not in string.punctuation) for s in file] # Removing punctuations
file = [s for s in file if s]

for word in file:
    word = ''.join(i for i in word if not i.isdigit()) #Removing unwanted numbers/digits attached with words
    opfile = word.split()
```

## Converting all the words on lower case

```python
A=[x.lower() for x in opfile] # List A has all the words from all files in lower case

N = len(A)                    # N is the total number of words in list A

print('\n','Total Number of Words present:',' - ', N,'\n')
```

## Calculating the frequency of each word in list A

```python
wordcount = {}                # Dictionary 'wordcount' is to store the each word and its frequency

for word in A:                # Calculating frequency by looping 
    if word not in wordcount:
        wordcount[word] = 1
    else:
        wordcount[word] +=1
            
#list(wordcount.keys())
#print(wordcount)
U = len(wordcount)            
print('\n','Total Number of Uniques Words present:',' - ', U,'\n')
```

## Sorting the frequencies in descending order of the word count using LAMBDA FUNCTION

```python
S = {}        #Empty dictionary to store the sorted word and their frequencies
S = sorted(wordcount.items(), key=lambda x: x[1], reverse=True)[:]    #Sorting in Descending order of key(Word Frequency)
for k,v in S:
    print(k," - ",v) #Printing each word and its frquency in descending order
```

## Defining Function for calculation the Probability of word 

```python
def prob(r):  # Defining function prob()
    return r/N
```

## Creating CSV file and loading the data & USE of FUNCTION

```python
with open('Data.csv', 'w') as csvfile:    # Creating a CSV file named 'Data' in write mode as csvfile
    fieldnames = ['WORD', 'FREQUENCY','RANK','PROBABILITY']  #Naming the title Row of the CSV
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    rank = 1                 # Initializing a variable 'rank' to 1 for ranking
    writer.writeheader()
    for k,v in S:
        if rank <= N:
            writer.writerow({'WORD': k,'FREQUENCY':v, 'RANK': rank,'PROBABILITY': prob(v)}) # Adding entry in the CSV
        rank += 1

#To Open and Display the csv file contents
        
#file = open('Data.csv','r')
#print(file.read())
```
