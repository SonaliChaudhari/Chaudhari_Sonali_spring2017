# Q1. Under package Gutenberg, see what files are in it.

- Find out how many words and how many unique words in file 'austen-persuasion.txt'.

- Calculate the frequency for each words and show the top 10 most frequent words and their frequency.

```python
import nltk
```

```python
nltk.corpus.gutenberg.fileids()

from nltk.corpus import gutenberg

S = nltk.corpus.gutenberg.words('austen-persuasion.txt')
```

```python
P = set([word.lower() for word in S])
```

```python
# Find out how many unique words in file 'austen-persuasion.txt'
# OPERATION USED : set(),word.lower(),len(P)

# ANSWER :
print("NUMBER OF UNIQUE WORDS IN FILE 'austen-persuasion.txt': ",len(P))
```

```python
# Calculate the frequency for each words
# ANSWER

# OPTION a:
wordcount={}

for word in gutenberg.words('austen-persuasion.txt'):
    if word not in wordcount:
        wordcount[word] = 1
    else:
        wordcount[word] += 1
list(wordcount.keys())
print(wordcount)
```

```python
# OPTION b:

for word,count in wordcount.items():
    print(word,"---",count,"\n")
```

```python
# Show the top 10 most frequent words and their frequency.
# ANSWER
print("\nTHE TOP 10 MOST FREQUENT WORDS:\n")
sorted(wordcount.items(), key=lambda x: x[1], reverse=True)[:10]
```
