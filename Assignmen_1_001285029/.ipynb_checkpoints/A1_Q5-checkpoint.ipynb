# Q.5 Use Gutenberg and Web_text data. 

- Find out what are the top 5 words that Shakespeare used but we do not use in currently.
- Take top 50 words from Shakespeare (all 3 books) and top 50 from Web_text (all the records).
- Remove punctuation and stop words.
- Remove the words we still use today, and get the unused list. Show the top 5 elements. 

```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
import string
```

```python
from nltk.corpus import webtext
webtext.fileids()
```

```python
from nltk.corpus import gutenberg
gutenberg.fileids()
```

```python

A = nltk.corpus.gutenberg.words('shakespeare-caesar.txt')
B = nltk.corpus.gutenberg.words('shakespeare-hamlet.txt')
C = nltk.corpus.gutenberg.words('shakespeare-macbeth.txt')
O = A+B+C

P = nltk.corpus.webtext.words('firefox.txt')
Q = nltk.corpus.webtext.words('overheard.txt')
R = nltk.corpus.webtext.words('grail.txt')
S = nltk.corpus.webtext.words('pirates.txt')
T = nltk.corpus.webtext.words('singles.txt')
U = nltk.corpus.webtext.words('wine.txt')
V = P+Q+R+S+T+U

# Removing punctuation AND stop words.


O=[''.join(c for c in s if c not in string.punctuation) for s in O]
O = [s for s in O if s]

V=[''.join(c for c in s if c not in string.punctuation) for s in V]
V = [s for s in V if s]

 
stop = stopwords.words('english')
S = [word for word in O if word not in stop]
W = [word for word in V if word not in stop]

S=[x.lower() for x in S]
W=[x.lower() for x in W]
```

```python
wordcount = {}

for word in S:
    if word not in W:
        if word not in wordcount:
            wordcount[word] = 1
        else:
            wordcount[word] +=1
            
list(wordcount.keys())
print(wordcount)  
```

```python
A = {}
A = sorted(wordcount.items(), key=lambda x: x[1], reverse=True)[:5]
print("\nTHE top 5 words that Shakespeare used but we do not use in currently.:\n\n",A)
print("\n---OR---\n")

for k,v in A:
    print(k)
```
