# Q.3 Under package Webtext, see what files are in it.

- Get content in file 'pirates.txt'.

- According to the number of lines for Jack Sparrow and Will Turner, find out who is the real leading man in this movie

```python
import nltk
```

```python
from nltk.corpus import webtext
webtext.fileids()
```

```python
# Get content in file 'pirates.txt'.
S = nltk.corpus.webtext.raw('pirates.txt')
P = set(S.split('\n'))
P
```

```python
#According to the number of lines for Jack Sparrow and Will Turner, find out who is the real leading man in this movie

js_count = 0
wt_count = 0

for line in P:
    if 'JACK SPARROW' in line:
        js_count += 1      
    if 'WILL TURNER' in line:
        wt_count += 1

print("Number of lines for JACK SPARROW: ",js_count,"\n\nNumber of lines for WILL TURNER:  ", wt_count,'\n')
print("Since Jack Sparrow has more line than Will Turner","\n\nJACK SPARROW IS THE LEADING MAN IN THE MOVIE.")    
```
